```python
import requests

url = "https://podatki.gov.si/dataset/ada88e06-14a2-49c4-8748-3311822e3585/resource/eb8b25ea-5c00-4817-a670-26e1023677c6/download/seznampraznikovindelaprostihdni20002030.csv"

response = requests.get(url)

#print(response.encoding)
response.encoding = "utf-8"
data = response.text
#print(data)

rezultat = {}
for vrstica in data.split("\r\n"):
    #print(vrstica)
    v_splitted = vrstica.split(";")
    if v_splitted[-1] == "2022":
        print(vrstica)
        
        dan = v_splitted[2]
        if dan in rezultat.keys():
            rezultat[dan] += 1
        else:
            rezultat[dan] = 1
            
print(rezultat)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_Game_of_Thrones_episodes"
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")
print(soup.prettify())
```

```python
soup_children = list(soup.children)
print(len(soup_children))
print(soup_children)
```

```python
html = soup_children[-1]
print(type(html))
print(html)
```

```python
print(html.attrs)
print(html.name)
```

```python
html_children = list(html.children)
for child in html_children:
    print(child.name)
```

```python
body = html_children[3]
print(body.name)
print(body.attrs)
```

```python
table = body.find("table", class_="wikitable plainrowheaders wikiepisodetable")
print(type(table))
print(table.name)
print(table.attrs)
print(table)
```

```python
for child in table.children:
    print(child.name)
```

```python
tbody = list(table.children)[0]

for child in tbody:
    print(child)
    print("\n\n\n")
```

```python
rows = list(tbody.children)[:2]

for row in rows:
    print(row.name)
    for column in row.children:
        print(column.text)
```

```python
tables = soup.find_all("table", class_="wikitable plainrowheaders wikiepisodetable")
print("Našlo smo ", len(tables), "tabel")
```

```python
import re
import matplotlib.pyplot as plt

episode_len = []

for table in tables[:2]:
    #print(table)
    rows = table.find_all("tr")
    for row in rows[1:]:
        tds = row.find_all("td")
        for td in tds:
            pass
        
        title_td = tds[1]
        a = title_td.find("a")
        href = a.attrs["href"]
        
        url2 = f"https://en.wikipedia.org{href}"
        print(url2)
        r2 = requests.get(url2)
        soup2 = BeautifulSoup(r2.text, "html.parser")
        
        infobox_table = soup2.find_all(class_="infobox")[0]
        #print(infobox_table)
        trs = infobox_table.find_all("tr")
        for tr in trs:
            th = tr.find("th")
            if th and th.text=="Running time":
                td = tr.find("td")
                print(td.text)
                match = re.search("\d+", td.text)
                runtime = int(match[0])
                episode_len.append(runtime)
                
plt.plot(range(len(episode_len)), episode_len)
plt.show()
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options


s = Service("./chromedriver101_0_4951_41")
chrome_options = Options()
chrome_options.add_argument("--headless")
with webdriver.Chrome(service=s, options=chrome_options) as driver:
    for moment in my_portfolio:
        driver.get("https://livetoken.co/listings/topshot")

        name_field = driver.find_element(by=By.CSS_SELECTOR, value="div.vs__selected-options input")
        #print(name_field.get_attribute("outerHTML"))
        WebDriverWait(webdriver, 10).until(EC.element_to_be_clickable(name_field))
        name_field.send_keys(moment["Name"])
        time.sleep(3)
        name_field.send_keys(Keys.ENTER)

        all_dropdowns = driver.find_elements(by=By.CSS_SELECTOR, value="div.vs__selected-options input")
        #print(all_dropdowns)
        date_dropdown = all_dropdowns[2]
        WebDriverWait(webdriver, 10).until(EC.element_to_be_clickable(date_dropdown))
        date_dropdown.send_keys(moment["Date"])
        time.sleep(6)
        date_dropdown.send_keys(Keys.ENTER)

        time.sleep(4)

        prices = driver.find_elements(by=By.CSS_SELECTOR, value="div.cost")
        price_1 = prices[0]
        price_2 = prices[1]
        #print(price_1.get_attribute("innerHTML"))
        #print(price_2.get_attribute("innerHTML"))
        price_1 = int(price_1.get_attribute("innerHTML").strip("$"))
        price_2 = int(price_2.get_attribute("innerHTML").strip("$"))
        my_price = (price_1 + price_2)/2
        print(f"My price for {moment['Name']} {moment['Date']} is ${my_price}")
```
